# Theory

**Q1. What does R-squared represent in a regression model?**

**Ans.** R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variables. It indicates how well the data fits the regression model, with values ranging from 0 to 1. A higher R-squared value means a better fit.


**Q2. What are the assumptions of linear regression?**

**Ans.** The assumptions of linear regression are:
1. **Linearity**: The relationship between the independent and dependent variables should be linear.
2. **Independence**: The residuals (errors) should be independent.
3. **Homoscedasticity**: The residuals should have constant variance at all levels of the independent variables.
4. **Normality**: The residuals should be normally distributed.



**Q3. What is the difference between R-squared and Adjusted R-squared?**

**Ans.** **R-squared** measures the proportion of variance in the dependent variable explained by the independent variables, indicating how well the model fits the data. However, R-squared can be artificially inflated by adding more predictors, even if they don't contribute meaningfully to the model. **Adjusted R-squared** addresses this issue by adjusting for the number of predictors, providing a more accurate measure of model fit by penalizing the inclusion of irrelevant variables. This makes Adjusted R-squared a more reliable indicator of a model's performance, especially when comparing models with different numbers of predictors.



**Q4. Why do we use Mean Squared Error (MSE)?**

**Ans.** Mean Squared Error (MSE) is used to measure the average of the squares of the errors. It indicates how close the fitted line is to the actual data points. A lower MSE indicates a better fit and helps in comparing different models.

**Q5. What does an Adjusted R-squared value of 0.85 indicate?**

**Ans.** An Adjusted R-squared value of 0.85 indicates that 85% of the variability in the dependent variable is explained by the independent variables in the model, after adjusting for the number of predictors. This suggests a strong model fit.


**Q6. How do we check for normality of residuals in linear regression?**

**Ans.** We check for normality of residuals using techniques like:
- Normal Probability Plot (Q-Q Plot)
- Histogram of residuals
- Shapiro-Wilk test
- Kolmogorov-Smirnov test






**Q7. What is multicollinearity, and how does it impact regression?**

**Ans.** Multicollinearity occurs when two or more independent variables in a regression model are highly correlated. It can inflate the variance of coefficient estimates, making them unstable and difficult to interpret. Methods to detect multicollinearity include Variance Inflation Factor (VIF) and correlation matrices.



**Q8. What is Mean Absolute Error (MAE)?**

**Ans.** Mean Absolute Error (MAE) is a measure of prediction accuracy in regression models. It represents the average of the absolute differences between predicted and actual values. MAE is easy to interpret and is less sensitive to outliers than MSE.



**Q9. What are the benefits of using an ML pipeline?**

**Ans.** The benefits of using an ML pipeline include:
- Automation of the workflow
- Improved reproducibility
- Easier experimentation with different models and parameters
- Efficient data preprocessing and feature selection
- Streamlined deployment and scaling



**Q10. Why is RMSE considered more interpretable than MSE?**

**Ans.** Root Mean Squared Error (RMSE) is considered more interpretable than MSE because it is in the same units as the dependent variable, making it easier to understand the magnitude of errors. RMSE penalizes larger errors more than MAE, providing a clearer picture of the model's performance.


**Q11. What is pickling in Python, and how is it useful in ML?**

**Ans.** Pickling in Python is the process of serializing and deserializing Python objects using the pickle module, which allows you to convert Python objects into a byte stream and save them to a file. This is particularly useful in machine learning, as it enables you to save trained models to disk and reload them later without the need to retrain the model. This can save time and computational resources, making it easier to deploy models in production and share them with others. However, it's important to be cautious when unpickling objects from untrusted sources, as this can pose security risks.



**Q12. What does a high R-squared value mean?**

**Ans.** A high R-squared value indicates that a large proportion of the variance in the dependent variable is explained by the independent variables. It suggests a good fit for the regression model.



**Q13. What happens if linear regression assumptions are violated?**

**Ans.** If linear regression assumptions are violated, the model's predictions and inferences may become unreliable. For instance, violating homoscedasticity can lead to biased standard errors, and non-normal residuals can affect confidence intervals and hypothesis tests.



**Q14. How can we address multicollinearity in regression?**

**Ans.** To address multicollinearity:
1. Remove highly correlated predictors
2. Use regularization techniques like Ridge or Lasso regression
3. Perform principal component analysis (PCA)
4. Center and standardize the predictors



**Q15. Why do we use pipelines in machine learning?**

**Ans.** Pipelines in machine learning automate the workflow, from data preprocessing to model training and evaluation. They improve reproducibility, streamline experimentation, and facilitate efficient scaling and deployment.



**Q16. How is Adjusted R-squared calculated?**

**Ans.** Adjusted R-squared is calculated using the formula:
$$\text{Adjusted } R^2 = 1 - \left( \frac{(1-R^2)(n-1)}{n-k-1} \right)$$
where \(R^2\) is the R-squared value, \(n\) is the number of observations, and \(k\) is the number of predictors.



**Q17. Why is MSE sensitive to outliers?**

**Ans.** MSE (Mean Squared Error) is sensitive to outliers because it squares the errors, which magnifies the impact of large deviations. As a result, a few extreme outliers can disproportionately affect the MSE value.



**Q18. What is the role of homoscedasticity in linear regression?**

**Ans.** Homoscedasticity means that the variance of the residuals is constant across all levels of the independent variables. It is crucial for ensuring unbiased and efficient estimates of the regression coefficients and reliable hypothesis tests.



**Q19. What is Root Mean Squared Error (RMSE)?**

**Ans.** RMSE (Root Mean Squared Error) is a measure of the average magnitude of the prediction errors. It is the square root of the mean of the squared errors and provides an interpretable metric in the same units as the dependent variable.



**Q20. Why is pickling considered risky?**

**Ans.** Pickling can be risky because it may execute arbitrary code during the deserialization process, making it vulnerable to security exploits if the pickle data comes from an untrusted source. Always validate and sanitize the input when using pickle.


**Q21. What alternatives exist to pickling for saving ML models?**

**Ans.** Alternatives to pickling for saving ML models include:
1. Joblib: Designed for serializing large data and models efficiently.
2. HDF5: Using libraries like h5py or PyTables for storing models in hierarchical data formats.
3. ONNX (Open Neural Network Exchange): A format for representing deep learning models.
4. PMML (Predictive Model Markup Language): An XML-based format for exporting models.



**Q22. What is heteroscedasticity, and why is it a problem?**

**Ans.** Heteroscedasticity occurs when the variance of the residuals is not constant across all levels of the independent variables. It is a problem because it can lead to inefficient and biased estimates of the regression coefficients, affecting the validity of statistical tests and confidence intervals.



**Q23. How does adding irrelevant predictors affect R-squared and Adjusted R-squared?**

**Ans.** Adding irrelevant predictors to a regression model can inflate the R-squared value, giving the false impression of a better fit. However, Adjusted R-squared accounts for the number of predictors and penalizes the inclusion of irrelevant variables, helping to provide a more accurate measure of model fit.

```python
# Q1. Write a Python script that calculates the Mean Squared Error (MSE) and Mean Absolute Error (MAE) for a multiple linear regression model using Seaborn's "diamonds" dataset.

import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Select features and target
X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate MSE and MAE
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
```

```python
# Q3. Write a Python script to check if the assumptions of linear regression are met. Use a scatter plot to check linearity, residuals plot for homoscedasticity, and correlation matrix for multicollinearity.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Example data
X = np.random.rand(100, 1)
y = 2 * X + 1 + np.random.randn(100, 1) * 0.2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Linearity check
plt.scatter(X_test, y_test, label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linearity Check')
plt.legend()
plt.show()

# Homoscedasticity check (Residual plot)
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.hlines(0, xmin=min(y_pred), xmax=max(y_pred), colors='red')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residuals Plot for Homoscedasticity Check')
plt.show()

# Multicollinearity check (Correlation matrix)
data = np.column_stack((X, y))
sns.heatmap(np.corrcoef(data.T), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Multicollinearity Check')
plt.show()
```

```python
# Q4. Create a machine learning pipeline that standardizes the features, fits a linear regression model, and evaluates the models R-squared score.

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# Example data
X = np.random.rand(100, 1)
y = 2 * X + 1 + np.random.randn(100, 1) * 0.2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create a pipeline with standardization and linear regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

# Fit the pipeline
pipeline.fit(X_train, y_train)

# Make predictions
y_pred = pipeline.predict(X_test)

# Evaluate the model's R-squared score
r2 = r2_score(y_test, y_pred)
print(f'R-squared score: {r2}')
```

```python
# Q21. Write a Python script to pickle a trained linear regression model, save it to a file, and load it back for prediction.

import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Generate synthetic data
X = np.random.rand(100, 1)
y = 2 * X + 1 + np.random.randn(100, 1) * 0.2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Pickle the model and save it to a file
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model has been pickled and saved to 'linear_regression_model.pkl'")

# Load the model from the file
with open('linear_regression_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

# Make predictions with the loaded model
y_pred = loaded_model.predict(X_test)
print(f'Predicted values: {y_pred[:5]}')
```

```python
# Q22. Write a Python script to perform linear regression with categorical features using one-hot encoding. Use the Seaborn 'tips' dataset.

import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the tips dataset
tips = sns.load_dataset('tips')

# Perform one-hot encoding on categorical features
tips_encoded = pd.get_dummies(tips, drop_first=True)

# Select features and target
X = tips_encoded.drop('tip', axis=1)
y = tips_encoded['tip']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Print coefficients, intercept, and R-squared score
print(f'Coefficients: {model.coef_}')
print(f'Intercept: {model.intercept_}')
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f'R-squared score: {r2}')
```

```python
# Q23. Compare Ridge Regression with Linear Regression on a synthetic dataset and print the coefficients and R-squared score.

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score

# Generate synthetic data
X = np.random.rand(100, 3)
y = 3 * X[:, 0] + 2 * X[:, 1] + X[:, 2] + 4 + np.random.randn(100) * 0.5

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
print(f'Linear Regression Coefficients: {linear_model.coef_}')
print(f'Linear Regression R-squared: {r2_linear}')

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f'Ridge Regression Coefficients: {ridge_model.coef_}')
print(f'Ridge Regression R-squared: {r2_ridge}')
```

```python
# Q24. Write a Python script that uses cross-validation to evaluate a Linear Regression model on a synthetic dataset.

import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression

# Generate synthetic data
X = np.random.rand(100, 3)
y = 3 * X[:, 0] + 2 * X[:, 1] + X[:, 2] + 4 + np.random.randn(100) * 0.5

# Create and fit the linear regression model
model = LinearRegression()

# Perform cross-validation
scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print(f'Cross-validated R-squared scores: {scores}')
print(f'Mean R-squared score: {np.mean(scores)}')
```

```python
# Q25. Write a Python script that compares polynomial regression models of different degrees and prints the R-squared score for each.

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

# Generate synthetic data
X = np.random.rand(100, 1)
y = 0.5 * X**3 - 2 * X**2 + X + np.random.randn(100, 1) * 0.2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Compare polynomial regression models of different degrees
degrees = [1, 2, 3, 4]
for degree in degrees:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f'Degree {degree} R-squared score: {r2}')
```
