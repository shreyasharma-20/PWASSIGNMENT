3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.


 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

```python
!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = bokeh.plotting.figure(x_range = fruits, title = "Fruit Counts", x_axis_label = "Fruits", y_axis_label = "Counts" , background_fill_color = 'orange')
p.vbar(x = fruits, top = counts, width = 0.5)
bokeh.plotting.show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data

. data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)

```python
import plotly.express as px
!pip install numpy
import numpy as np

!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = bokeh.plotting.figure(title = "Histogram", x_axis_label = "Value", y_axis_label = "Frequency")
p.quad(top = hist, bottom = 0, left = edges[:-1], right = edges[1:], fill_color = "pink", line_color = "purple")
bokeh.plotting.show(p)
```

5.Create a Bokeh heatmap using the provided dataset.

 data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)

```python
import plotly.express as px
!pip install numpy
import numpy as np

!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = bokeh.plotting.figure(title = "Heatmap", x_axis_label = "X-axis", y_axis_label = "Y-axis")
p.image(image = [data_heatmap], x = [xx.min()], y = [yy.min()], dw = [xx.ptp()], dh = [yy.ptp()], palette = "Spectral11")
bokeh.plotting.show(p)
```
