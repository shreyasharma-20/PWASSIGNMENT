 ### **(iv) Write a Python function to find the geometric mean of a list of positive numbers.**

````python
import numpy as np
import scipy.stats as stats

# Function to calculate the geometric mean
def calculate_geometric_mean(values):
    """
    Calculate the geometric mean of a list of positive numbers.

    Parameters:
    values (list): List of positive numerical values.

    Returns:
    float: The geometric mean of the values.
    """
    if any(v <= 0 for v in values):  # Check if all values are positive
        raise ValueError("All values must be positive.")

    # Calculate the geometric mean using scipy.stats.gmean()
    return stats.gmean(values)

# Example usage
int_list = np.random.randint(90, 131, size=100).tolist()  # Generate 100 random integers between 90 and 130
geometric_mean_value = calculate_geometric_mean(int_list)

print("Geometric Mean:", geometric_mean_value)
````

### **(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).**

````python
import numpy as np

# Function to calculate the midrange
def calculate_midrange(values):
    """
    Calculate the midrange of a list of numbers (average of min and max).

    Parameters:
    values (list): List of numerical values.

    Returns:
    float: The midrange of the values.
    """
    # Ensure the list is not empty
    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    # Calculate the midrange (average of min and max values)
    min_value = np.min(values)
    max_value = np.max(values)
    midrange_value = (min_value + max_value) / 2
    return midrange_value

# Example usage
int_list = np.random.randint(90, 131, size=100).tolist()  # Generate 100 random integers between 90 and 130
midrange_value = calculate_midrange(int_list)

print("Midrange:", midrange_value)
````

  ### **(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.**

````python
import numpy as np

# Function to calculate the trimmed mean
def calculate_trimmed_mean(values, trim_percentage):
    """
    Calculate the trimmed mean of a list of numbers by excluding a certain percentage of outliers.

    Parameters:
    values (list): List of numerical values.
    trim_percentage (float): Percentage of values to remove from both ends.

    Returns:
    float: The trimmed mean of the values.
    """
    if not 0 <= trim_percentage <= 50:
        raise ValueError("Trim percentage must be between 0 and 50.")

    if len(values) == 0:
        raise ValueError("The list cannot be empty.")

    # Sort the list of values
    sorted_values = np.sort(values)

    # Calculate the number of elements to trim
    num_trim = int(len(sorted_values) * (trim_percentage / 100))

    # Remove the trim_percentage of smallest and largest values
    trimmed_values = sorted_values[num_trim:len(sorted_values) - num_trim]

    # Calculate and return the mean of the remaining values
    trimmed_mean = np.mean(trimmed_values)
    return trimmed_mean

# Example usage
int_list = np.random.randint(90, 131, size=100).tolist()  # Generate 100 random integers between 90 and 130

# Calculate the trimmed mean with 10% trimming from both ends
trimmed_mean_value = calculate_trimmed_mean(int_list, 10)

print("Trimmed Mean (10%):", trimmed_mean_value)
````

# **5. Create a Python function to generate random samples from a given probability distribution (e.g.,binomial, Poisson) and calculate their mean and variance.**

````python
import numpy as np

def generate_samples(distribution_type, **params):
    """
    Generate random samples from a given probability distribution.

    Parameters:
    distribution_type (str): The type of distribution ('binomial' or 'poisson').
    **params: The parameters required for the distribution.

    Returns:
    samples (np.ndarray): Array of generated random samples.
    mean (float): Mean of the samples.
    variance (float): Variance of the samples.
    """
    if distribution_type == 'binomial':
        # Ensure both n (number of trials) and p (probability of success) are provided
        n = params.get('n')
        p = params.get('p')
        size = params.get('size', 1000)  # Default to 1000 samples if not specified

        # Check if n or p are None
        if n is None or p is None:
            raise ValueError("Both 'n' (trials) and 'p' (probability) must be specified for binomial distribution.")

        samples = np.random.binomial(n, p, size)

    elif distribution_type == 'poisson':
        # Ensure lambda (rate of events) is provided
        lam = params.get('lambda_')  # Use 'lambda_' because 'lambda' is a reserved keyword
        size = params.get('size', 1000)  # Default to 1000 samples if not specified

        # Check if lambda_ is None
        if lam is None:
            raise ValueError("The 'lambda_' (rate of events) must be specified for poisson distribution.")

        samples = np.random.poisson(lam, size)

    else:
        raise ValueError("Invalid distribution type. Use 'binomial' or 'poisson'.")

    # Calculate mean and variance of the generated samples
    mean = np.mean(samples)
    variance = np.var(samples)

    return samples, mean, variance

# Example usage
try:
    # Binomial distribution with 10 trials and probability 0.5 (like flipping a fair coin)
    binomial_samples, binomial_mean, binomial_variance = generate_samples(
        'binomial', n=10, p=0.5, size=10000)

    # Poisson distribution with lambda = 5 (rate of 5 events)
    poisson_samples, poisson_mean, poisson_variance = generate_samples(
        'poisson', lambda_=5, size=10000)

    print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")
    print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")

except ValueError as e:
    print(f"Error: {e}")
````

# **6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.**

````python
import numpy as np

def generate_gaussian_samples(mean, std_dev, size):
    """
    Generate random samples from a Gaussian (normal) distribution.

    Parameters:
    mean (float): The mean of the Gaussian distribution.
    std_dev (float): The standard deviation of the Gaussian distribution.
    size (int): The number of samples to generate.

    Returns:
    np.ndarray: Array of generated random samples.
    float: Mean of the samples.
    float: Variance of the samples.
    float: Standard deviation of the samples.
    """
    # Generate random samples from Gaussian distribution
    samples = np.random.normal(mean, std_dev, size)

    # Calculate mean, variance, and standard deviation of the samples
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)

    return samples, sample_mean, sample_variance, sample_std_dev

# Example usage
mean = 50        # Mean of the distribution
std_dev = 10     # Standard deviation of the distribution
size = 1000      # Number of samples to generate

# Generate samples and calculate statistics
samples, sample_mean, sample_variance, sample_std_dev = generate_gaussian_samples(mean, std_dev, size)

# Print the results
print(f"Generated {size} random samples from a Gaussian distribution")
print(f"Mean of the samples: {sample_mean}")
print(f"Variance of the samples: {sample_variance}")
print(f"Standard Deviation of the samples: {sample_std_dev}")
````

#**Que 8 Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.**


````python
import math

def normal_pdf(x, mu, sigma):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    x (float): The point at which the PDF is evaluated.
    mu (float): The mean (mu) of the normal distribution.
    sigma (float): The standard deviation (sigma) of the normal distribution.

    Returns:
    float: The value of the PDF at x.
    """
    # Calculate the PDF using the formula for normal distribution
    coefficient = 1 / (sigma * math.sqrt(2 * math.pi))
    exponent = math.exp(-0.5 * ((x - mu) / sigma) ** 2)

    return coefficient * exponent

# Example usage
x = 0
mu = 0
sigma = 1

pdf_value = normal_pdf(x, mu, sigma)
print(f"The PDF at x = {x} for N({mu}, {sigma}) is {pdf_value}")
````

# **Que 9 Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.**

````python
import math

def exponential_cdf(x, lambda_param):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    Parameters:
    x (float): The value at which the CDF is evaluated.
    lambda_param (float): The rate parameter (lambda) of the exponential distribution.

    Returns:
    float: The value of the CDF at x.
    """
    if x < 0:
        return 0.0  # CDF is 0 for x < 0
    return 1 - math.exp(-lambda_param * x)

# Example usage
x = 2
lambda_param = 1.5

cdf_value = exponential_cdf(x, lambda_param)
print(f"The CDF at x = {x} for Exp({lambda_param}) is {cdf_value}")
````

# **Que 10 Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.**

````python
import math

def poisson_pmf(k, lambda_param):
    """
    Calculate the Probability Mass Function (PMF) of a Poisson distribution.

    Parameters:
    k (int): The number of events (k) for which the PMF is evaluated.
    lambda_param (float): The rate parameter (lambda) of the Poisson distribution.

    Returns:
    float: The probability of observing exactly k events.
    """
    if k < 0 or lambda_param <= 0:
        raise ValueError("k must be non-negative and lambda must be positive.")

    # Calculate the Poisson PMF using the formula
    pmf = (lambda_param ** k) * math.exp(-lambda_param) / math.factorial(k)
    return pmf

# Example usage
k = 3
lambda_param = 2.5

pmf_value = poisson_pmf(k, lambda_param)
print(f"The PMF at k = {k} for Poisson({lambda_param}) is {pmf_value}")
````

````python
import numpy as np
import scipy.stats as stats

# Data for the old and new layouts
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases, 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases, 1000 visitors

# Calculate the sample proportions
p1 = np.mean(old_layout)  # Proportion of purchases in old layout
p2 = np.mean(new_layout)  # Proportion of purchases in new layout

# Number of successes (purchases) and sample sizes
x1 = np.sum(old_layout)  # Successes in old layout
x2 = np.sum(new_layout)  # Successes in new layout
n1 = len(old_layout)     # Sample size for old layout
n2 = len(new_layout)     # Sample size for new layout

# Pooled proportion
p_hat = (x1 + x2) / (n1 + n2)

# Z-statistic calculation
z = (p1 - p2) / np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))  # Two-tailed test

# Set the significance level
alpha = 0.05

# Print the results
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in conversion rates.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in conversion rates.")
````
