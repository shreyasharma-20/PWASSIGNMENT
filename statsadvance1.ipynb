Probability Distribution
A probability distribution is a mathematical function that provides the probabilities of occurrence of different possible outcomes in a random experiment. It describes how the probabilities are distributed over the possible values of a random variable. Probability distributions can be classified into two main categories:

Discrete Probability Distributions: These are used for discrete random variables, which can take on a countable number of values (e.g., the number of heads in coin flips). Examples include the binomial distribution and the Poisson distribution.

Continuous Probability Distributions: These apply to continuous random variables, which can take on an infinite number of values within a given range (e.g., the height of individuals). Examples include the normal distribution and the exponential distribution.

Probability Density Function (PDF)
A probability density function (PDF) is a function associated with continuous random variables that describes the likelihood of the random variable taking on a particular value. The PDF itself does not give probabilities directly, but rather provides a way to calculate probabilities over intervals.

# Question7: Explain the Central Limit Theorem (CLT) with example.

The Central Limit Theorem (CLT) is a fundamental statistical concept that states that, given a sufficiently large sample size, the sampling distribution of the sample mean will approximate a normal distribution, regardless of the original distribution of the population. This holds true as long as the samples are independent and identically distributed.

Key Points of the CLT:
Sample Size: Generally, a sample size of 30 or more is considered sufficient for the CLT to apply, though this can vary depending on the shape of the population distribution.
Independence: The samples must be drawn independently from the population.
Normal Distribution: As the sample size increases, the distribution of the sample means will tend to be normal, even if the underlying population distribution is not.
Example:
Suppose you want to analyze the average height of adult men in a city, where the heights are not normally distributed (e.g., they could be skewed to one side).

Population: Let's say the average height is 175 cm with a standard deviation of 10 cm.
Sampling: You take random samples of size 30 from this population.
Sample Means: For each sample, you calculate the mean height. You do this many times (e.g., 1000 samples).
Distribution of Sample Means: According to the CLT, if you were to plot the distribution of these sample means, it would approximate a normal distribution with:
Mean equal to the population mean (175 cm).
Standard deviation equal to the population standard deviation divided by the square root of the sample size (10/‚àö30 ‚âà 1.83 cm).
As a result, even though the original height data might be skewed, the distribution of the sample means will be approximately normal, allowing you to make inferences about the population mean using techniques that assume normality.

This property of the CLT is crucial for hypothesis testing and confidence intervals in statistics.

```python
#Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
#size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to
#reject the null hypothesis?

 #Task: Write Python code to calculate the z-score and p-value for the given data.

#Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.


import numpy as np
import scipy.stats as stats

# Given data
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25

# Calculate the z-score
z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    decision = "Reject the null hypothesis."
else:
    decision = "Fail to reject the null hypothesis."

# Output results
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(decision)
```

Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.
Generate 1,000 samples and plot the distribution. What is the expected mean and variance?

Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.

Objective: Understand the properties of a binomial distribution and verify them through simulation



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
n = 10  # number of trials
p = 0.6  # probability of success
num_samples = 1000  # number of samples

# Generate samples from a binomial distribution
samples = np.random.binomial(n, p, num_samples)

# Calculate mean and variance
sample_mean = np.mean(samples)
sample_variance = np.var(samples)

# Calculate expected mean and variance
expected_mean = n * p
expected_variance = n * p * (1 - p)

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.histplot(samples, bins=np.arange(-0.5, n+1.5, 1), kde=False, stat="probability")
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(np.arange(n+1))
plt.grid(axis='y')

# Show expected mean and variance on the plot
plt.axvline(expected_mean, color='r', linestyle='dashed', linewidth=2, label=f'Expected Mean: {expected_mean:.2f}')
plt.axvline(sample_mean, color='g', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(expected_mean + np.sqrt(expected_variance), color='orange', linestyle='dashed', linewidth=2, label=f'Expected Variance: {expected_variance:.2f}')
plt.axvline(expected_mean - np.sqrt(expected_variance), color='orange', linestyle='dashed', linewidth=2)

plt.legend()
plt.show()

# Print calculated mean and variance
print(f'Sample Mean: {sample_mean:.2f}')
print(f'Sample Variance: {sample_variance:.2f}')
print(f'Expected Mean: {expected_mean:.2f}')
print(f'Expected Variance: {expected_variance:.2f}')
```
